```python
# 1. Import necessary libraries (matplotlib and seaborn). 
#    Load the csv file of dataset into dataframe in Python program. [1 pt]
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('us_weather_history.csv')
weather_df = pd.DataFrame(data)
weather_df.head()
```

```python
# 2. Plot min and max for actual_mean_temp, actual_min_temp and actual_max_temp using bar graph. 
#    Use blue for min and red for max. [1 pt]
x = ['actual_mean_temp', 'actual_min_temp', 'actual_max_temp']
y_max = [weather_df['actual_mean_temp'].max(), weather_df['actual_min_temp'].max(), weather_df['actual_max_temp'].max()]
y_min = [weather_df['actual_mean_temp'].min(), weather_df['actual_min_temp'].min(), weather_df['actual_max_temp'].min()]

plt.bar(x, y_min, color = 'blue')
plt.bar(x, y_max, bottom = y_min, color = 'red')
plt.title('Min and Max : Actual Mean Temp, Actual Min Temp, Actual Max Temp')
plt.show()

# Observation : 
#       actual_mean_temp   actual_min_temp   actual_max_temp
# min          18                 7                  26
# max          88                 75                100
```

```python
# 3. Plot two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015.
#    Make key observations (put them as comments in jupyter notebook). [1 pt]

# Aug 2014
index = weather_df.index
ll1 = index[weather_df['date'] == '2014-8-1'].tolist()
ll1 = ll1[0]
ul1 = index[weather_df['date'] == '2014-8-31'].tolist()
ul1 = ul1[0]

x = weather_df['actual_mean_temp'][ll1:ul1 + 1]
plt.hist(x, 7, facecolor = 'yellow', alpha = 0.5, edgecolor = 'black')
plt.ylabel('Number of Days')
plt.xlabel('Temperature')
plt.title('Number of Days vs Actual Mean Temp for Aug 2014')
plt.show()

# Observations : 
# 1. August 2014 recorded 10 days with temperature between 76.2 and 78.4
# 2. It recorded equal number of days with temperature between '72.1 - 74.1 and 74.2 - 76.2 (2)' and '78.4 - 80.4 and 78.5 - 83 (5)'
```

```python
# March 2015
index = weather_df.index
ll1 = index[weather_df['date'] == '2015-3-1'].tolist()
ll1 = ll1[0]
ul1 = index[weather_df['date'] == '2015-3-31'].tolist()
ul1 = ul1[0]

x = weather_df['actual_mean_temp'][ll1:ul1 + 1]
plt.hist(x, 7, facecolor = 'green', alpha = 0.5, edgecolor = 'black')
plt.ylabel('Number of Days')
plt.xlabel('Temperature')
plt.title('Number of Days vs Actual Mean Temp for Mar 2015')
plt.show()

# Observation : 
# 1. It recorded equal number of days with temperature between '50.1 - 55.1 and 60.1 - 65.1 (7)' and '35 - 40 and 65.2 - 70.2'
# 2. Upto temperature of 50, as the temperture rises, the number of days recorded increases
```

```python
#4. Plot box plots for attributes actual_mean_temp, actual_min_temp and actual_max_temp. 
#   Make key observations (put them as comments in jupyter notebook). [2 pts]

boxplot = weather_df.boxplot(column = ['actual_mean_temp', 'actual_min_temp', 'actual_max_temp'], grid = False, figsize = (5,5))
boxplot

# Observations
#                actual_mean_temp   actual_min_temp   actual_max_temp
# 1st Quartile          47                 37                 58
# median                63                 52                 73
# 3rd Quartile          75                 65                 86
```

```python
# 5. Find the correlation between the following pairs of attributes: 
#    actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. 
#    Make key observations. [1 pt]
corr_1 = weather_df['actual_mean_temp'].corr(weather_df['actual_min_temp'], method = 'pearson')
corr_2 = weather_df['actual_mean_temp'].corr(weather_df['actual_max_temp'], method = 'pearson')
print("Correlation between 'actual_mean_temp' & 'actual_min_temp' : ", corr_1)
print("Correlation between 'actual_mean_temp' & 'actual_max_temp' : ", corr_2)

# Observation: Both values of correlation are very high (~0.97).
```

```python
# 6. Plot two scatter plots between the following pairs of attributes:
#    actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. 
#    Make key observations. 

x1 = weather_df['actual_mean_temp']
y1 = weather_df['actual_min_temp']
plt.scatter(x1, y1)
plt.xlabel('Actual Mean Temperature')
plt.ylabel('Actual Min Temperature')
plt.title("Relation between 'Actual Mean Temp' and 'Actual Min Temp'")
plt.show()

# Observation : Actual mean temperature and actual min temperature follow an almost linear relationship
```

```python
x2 = weather_df['actual_mean_temp']
y2 = weather_df['actual_max_temp']
plt.scatter(x2, y2, color = 'purple')
plt.xlabel('Actual Mean Temperature')
plt.ylabel('Actual Max Temperature')
plt.title("Relation between 'Actual Mean Temp' and 'Actual Max Temp'")
plt.show()

# Observation : actual mean temperature and actual max temperature follow an almost linear relationship
```

```python
# 7.  Make correlation matrix plot and scatterplot matrix taking all temperature attributes in the dataset. [2 pts] 
correlations = weather_df.corr()

sns.heatmap(correlations, annot = False, square = True)
plt.show()

# Key Observation : 
#1. The map is symmetrical along the diagonal, ie bottom left and top right are identical
```
