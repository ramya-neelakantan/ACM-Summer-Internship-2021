# Question 1 
Create your Twitter developer account. Write python code (use twython / tweepy library) to : 
    1.1 Collect last 1000 tweets posted by ‘narendra modi’ 
    1.2 Collect 2000 tweets containing keyword ‘covid’. [5 pt] 

```python
import tweepy
from tweepy import * 
import pandas as pd
```

```python
# Authentication using Twitter access keys and secret keys
def authentication():
    consumer_key = "" 
    consumer_secret = ""
    access_key = ""
    access_secret = ""
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_key, access_secret)
    api = tweepy.API(auth, wait_on_rate_limit = True)
    return api
```

```python
# 1.1
import tweepy

file = open("tweets_file_1.txt", "w+")

def get_tweets(username):
    i = 0
    number_of_tweets = 1000
    api = authentication()
    for tweet in tweepy.Cursor(api.user_timeline, screen_name = username, 
                                    tweet_mode="extended").items(number_of_tweets):
        #print(tweet.full_text) 
        file.write(tweet.full_text)
        i += 1       
    file.close()
    print("Total Tweets : ", i)

if __name__ == '__main__':
    username = input("Enter Username : ")
    get_tweets(username) 
```

```python
# 1.2

file1 = open('tweets_file_2.txt', 'w+')

def get_covid_tweets() : 
    i = 0
    number_of_tweets = 2000
    api = authentication()
    search_word = "covid" 
    new_search = search_word + " -filter:retweets"
    for tweet in tweepy.Cursor(api.search, q = new_search, lang="en").items(number_of_tweets):
        #print(tweet.text)
        file1.write(tweet.text)
        i += 1
    file1.close()
    print("Total Tweets : ", i)
    
if __name__ == '__main__' : 
    get_covid_tweets()
```

# Question 2 
Write a python program (use web scraping concept, use beautiful soup library or any other) to collect news title, author, date, time, and details for all news items that appears on a single page of https://inshorts.com/en/read/ Store each news items details in a dictionary, and create a list of such dictionary objects, each object representing a news item. Save these dictionary objects in a file, each object representing a news item on a separate line. [5 pt]

```python
# Solution : 
```

```python
from urllib.request import urlopen
```

```python
# Open URL of InShorts and get http response object
ins_url = "https://inshorts.com/en/read/"
ins_data = urlopen(ins_url)
print(type(ins_data))
```

```python
# Read HTML obtained from website
ins_html = ins_data.read()
ins_data.close()
#ins_html
```

```python
# Parse the HTML file using BeautifulSoup
from bs4 import BeautifulSoup as soup
import parser
ins_soup = soup(ins_html, 'html.parser')
```

```python
# Create a list of all titles in the website
titles = ins_soup.find_all('span', {}, itemprop = 'description')
titles_final = []
for i in range(0, len(titles)) : 
    titles_final.append(titles[i].get('content'))
titles_final
```

```python
# Create a list of all authors in the website
authors = ins_soup.find_all('span', {'itemprop' : 'name'})
authors_final = []
for i in range(0, len(authors)) : 
    if(authors[i].get('content') == 'Inshorts') : 
        continue
    else : 
        authors_final.append(authors[i].get('content'))
authors_final
```

```python
# Create a list of all dates and corresponding day in the website
dates = ins_soup.find_all('span', {'clas' : 'date'})
dates_final = []
for i in range(0, len(dates)):
    dates_final.append(dates[i].get_text())
dates_final
```

```python
# Create a list of all times at which news was generated in the website
times = ins_soup.find_all('span', {'class' : 'time'}, itemprop = 'datePublished')
times_final = []
for i in range(0, len(times)) : 
    times_final.append(times[i].get_text())
times_final
```

```python
# Create a list of all news articles in the website
news_articles = []
main_news = ins_soup.find_all('div', {}, itemprop = 'articleBody')
for i in range(0, len(main_news)):
    news_articles.append(main_news[i].get_text())
print(news_articles) 
```

```python
# First, create dictionary items containing Title, Author, Date and Day, Time and News article for each news item
# Second, append all dictionaries to a list
list_of_dict = []
for i in range(0, len(news_articles)) : 
  dict_news_article = {'Title': titles_final[i], 'Author': authors_final[i], 'Date': dates_final[i], 
           'Time' : times_final[i], 'News Article' : news_articles[i]}
  list_of_dict.append(dict_news_article)

list_of_dict
```

```python
# Method 2 : Create a dictionary with Title, Author, Date and Day, Time and News article for all news items
dict = {'Title': titles_final, 'Author': authors_final, 'Date': dates_final, 'Time' : times_final, 'News Article' : news_articles}
dict
```

```python
# Convert dictionary created in earlier step into DataFrame for tabular form of data
ins_df = pd.DataFrame(dict)
ins_df
```

```python
# Create csv file to store the data in tabular form 
ins_df.to_csv('ins_data.csv', index = False)
```
